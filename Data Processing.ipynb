```python
import requests
import pandas as pd
from datetime import datetime

# Функция для получения данных по погоде с NASA POWER
def get_nasa_power_data(lat, lon, start_date, end_date):
    url = "https://power.larc.nasa.gov/api/temporal/daily/point"
    params = {
        'start': start_date.strftime('%Y%m%d'),
        'end': end_date.strftime('%Y%m%d'),
        'latitude': lat,
        'longitude': lon,
        'community': 'AG',
        'parameters': 'T2M,PRECTOTCORR,PS,QV2M',
        'format': 'json',
        'header': 'true',
        'time-standard': 'UTC'
    }
    
    response = requests.get(url, params=params)
    
    if response.status_code == 200:
        data = response.json()['properties']['parameter']
        
        # Преобразуем данные в удобный формат
        weather_data = []
        for date in data['T2M']:
            weather_data.append({
                'lat': lat,
                'lon': lon,
                'date': date,
                'temp': data['T2M'][date],
                'precipitation': data['PRECTOTCORR'][date],
                'pressure': data['PS'][date],
                'humidity': data['QV2M'][date]
            })
        
        return weather_data
    else:
        print(f"Error: {response.status_code}")
        return None

'''# Загрузка данных для заданных координат и периода
lat = 55.7558
lon = 37.6176
start_date = datetime(1998, 1, 1)
end_date = datetime(2018, 12, 31)

weather_data = get_nasa_power_data(lat, lon, start_date, end_date)

# Преобразование в DataFrame
if weather_data:
    weather_df = pd.DataFrame(weather_data)
    
    # Сохранение данных в CSV
    weather_df.to_csv('nasa_power_weather_data.csv', index=False)
    
    print(weather_df.head())
'''
```

```python
data = data.loc[3810:4692]
import os
# Создание папки для хранения климатических данных
if not os.path.exists('climate'):
    os.makedirs('climate')

# Параметры временного периода
start_date = datetime(1998, 1, 1)
end_date = datetime(2018, 12, 31)

# Обработка каждой координаты и сохранение данных в отдельный файл
for idx, row in data.iterrows():
    lat = row['latitude']#longitude	latitude
    lon = row['longitude']
    weather_data = get_nasa_power_data(lat, lon, start_date, end_date)
    
    if weather_data:
        weather_df = pd.DataFrame(weather_data)
        file_name = f'climate/{idx+1}.csv'  # Нумерация файлов начинается с 1
        weather_df.to_csv(file_name, index=False)
        print(f'Saved data for lat: {lat}, lon: {lon} to {file_name}')
    else:
        print(f'Failed to get data for lat: {lat}, lon: {lon}')
```
