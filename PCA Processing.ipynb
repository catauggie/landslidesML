```python
import requests
import pandas as pd

def get_soil_composition(lat, lon):
    url = f"https://rest.isric.org/soilgrids/v2.0/properties/query?lon={lon}&lat={lat}"
    response = requests.get(url)
    
    if response.status_code == 200:
        data = response.json()
        properties = data.get('properties', {})
        
        soil_data = {
            'sand': properties.get('sand', {}).get('mean', None),
            'silt': properties.get('silt', {}).get('mean', None),
            'clay': properties.get('clay', {}).get('mean', None),
            'organic_carbon': properties.get('ocd', {}).get('mean', None)
        }
        return soil_data
    else:
        print(f"Error fetching data for coordinates ({lat}, {lon}): {response.status_code}")
        return {'sand': None, 'silt': None, 'clay': None, 'organic_carbon': None}

# Example usage
lat = 40.7128
lon = -74.0060
soil_composition = get_soil_composition(lat, lon)
print(f'Soil composition at ({lat}, {lon}): {soil_composition}')

# Generalize for all coordinates in data
def add_soil_composition_to_dataframe(df):
    df['sand'] = None
    df['silt'] = None
    df['clay'] = None
    df['organic_carbon'] = None

    for index, row in df.iterrows():
        soil_composition = get_soil_composition(row['latitude'], row['longitude'])
        df.at[index, 'sand'] = soil_composition['sand']
        df.at[index, 'silt'] = soil_composition['silt']
        df.at[index, 'clay'] = soil_composition['clay']
        df.at[index, 'organic_carbon'] = soil_composition['organic_carbon']

    return df

# Example DataFrame
data = {
    'latitude': [40.7128, 34.0522],
    'longitude': [-74.0060, -118.2437]
}
df = pd.DataFrame(data)

# Add soil composition data
df = add_soil_composition_to_dataframe(df)

# Save to CSV
#df.to_csv('data_with_soil_composition.csv', index=False)
df
```

```python
import requests
import pandas as pd

def get_soil_composition(lat, lon):
    url = f"https://rest.isric.org/soilgrids/v2.0/properties/query?lon={lon}&lat={lat}"
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        # Парсинг данных о составе почвы
        soil_data = {
            'sand': data['properties']['sand']['mean'] if 'sand' in data['properties'] else None,
            'silt': data['properties']['silt']['mean'] if 'silt' in data['properties'] else None,
            'clay': data['properties']['clay']['mean'] if 'clay' in data['properties'] else None,
            'organic_carbon': data['properties']['ocd']['mean'] if 'ocd' in data['properties'] else None
        }
        return soil_data
    else:
        return {'sand': None, 'silt': None, 'clay': None, 'organic_carbon': None}

# Пример использования
lat = 40.7128
lon = -74.0060
soil_composition = get_soil_composition(lat, lon)
print(f'Soil composition at ({lat}, {lon}): {soil_composition}')
```
