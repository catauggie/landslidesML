```python
import os
import pandas as pd
from pathlib import Path

# Путь к папке с исходными файлами
input_folder = 'climate'

# Путь к папке для сохранения новых файлов
output_folder = 'climate_new'

# Создаем папку для новых файлов, если её еще нет
Path(output_folder).mkdir(parents=True, exist_ok=True)

# Функция для чтения данных из файла, агрегации по месяцам и сохранения в новый файл
def process_file(file_path):
    # Чтение файла с данными погоды
    df = pd.read_csv(file_path)
    
    # Извлечение числа из имени файла
    filename = os.path.splitext(os.path.basename(file_path))[0]
    number_from_filename = filename.split('_')[0]  # предполагаем, что число в начале имени файла
    
    # Преобразование столбца с датой к типу datetime
    df['date'] = pd.to_datetime(df['date'], format='%Y%m%d')
    
    # Извлечение года и месяца из даты
    df['year_month'] = df['date'].dt.to_period('M')
    
    # Группировка и усреднение по месяцам
    monthly_data = df.groupby('year_month').agg({
        'temp': 'mean',
        'precipitation': 'mean',
        'pressure': 'mean',
        'humidity': 'mean'
    }).reset_index()
    
    # Путь для сохранения нового файла
    output_file = os.path.join(output_folder, f'{number_from_filename}.csv')
    
    # Сохранение в новый CSV файл
    monthly_data.to_csv(output_file, index=False)
    print(f'Создан файл: {output_file}')

# Получаем список файлов в папке climate
files = os.listdir(input_folder)

# Обрабатываем каждый файл
for file in files:
    file_path = os.path.join(input_folder, file)
    process_file(file_path)

print('Все файлы обработаны и сохранены в папке climate_new.')
```

```python
import os
import pandas as pd
from pathlib import Path

# Путь к папке с исходными файлами
input_folder = 'climate'

# Путь к папке для сохранения новых файлов
output_folder = 'climate_new'

# Создаем папку для новых файлов, если её еще нет
Path(output_folder).mkdir(parents=True, exist_ok=True)

# Функция для чтения данных из файла, агрегации по месяцам и сохранения в новый файл
def process_file(file_path):
    # Чтение файла с данными погоды
    df = pd.read_csv(file_path)
    
    # Извлечение координат из первой строки файла
    lat = df['lat'].iloc[0]
    lon = df['lon'].iloc[0]
    
    # Преобразование столбца с датой к типу datetime
    df['date'] = pd.to_datetime(df['date'], format='%Y%m%d')
    
    # Извлечение года и месяца из даты
    df['year_month'] = df['date'].dt.to_period('M')
    
    # Группировка и усреднение по месяцам
    monthly_data = df.groupby('year_month').agg({
        'temp': 'mean',
        'precipitation': 'mean',
        'pressure': 'mean',
        'humidity': 'mean'
    }).reset_index()
    
    # Добавление координат в DataFrame
    monthly_data['lat'] = lat
    monthly_data['lon'] = lon
    
    # Путь для сохранения нового файла
    output_file = os.path.join(output_folder, f'{lat}_{lon}_monthly.csv')
    
    # Сохранение в новый CSV файл
    monthly_data.to_csv(output_file, index=False)
    print(f'Создан файл: {output_file}')

# Получаем список файлов в папке climate
files = os.listdir(input_folder)

# Обрабатываем каждый файл
for file in files:
    file_path = os.path.join(input_folder, file)
    process_file(file_path)

print('Все файлы обработаны и сохранены в папке climate_new.')
```

```python
import os
import pandas as pd

# Путь к папке с основными файлами
input_folder = 'climate_new'

# Чтение основного DataFrame
main_df = pd.read_excel('data.xlsx')
#main_df = main_df.loc[0:10]

# Функция для добавления фичей по месяцам из дополнительного файла
def add_monthly_features(row):
    number = row['number']
    filename = os.path.join(input_folder, f'{number}.csv')
    
    if os.path.exists(filename):
        # Чтение дополнительного файла
        additional_data = pd.read_csv(filename)
        
        # Преобразование столбца с датой к типу datetime
        additional_data['year_month'] = pd.to_datetime(additional_data['year_month'])
        
        # Обработка каждого значения по месяцам
        for index, month_data in additional_data.iterrows():
            year_month = month_data['year_month'].strftime('%Y-%m')
            main_df.loc[row.name, f'temp_{year_month}'] = month_data['temp']
            main_df.loc[row.name, f'precipitation_{year_month}'] = month_data['precipitation']
            main_df.loc[row.name, f'pressure_{year_month}'] = month_data['pressure']
            main_df.loc[row.name, f'humidity_{year_month}'] = month_data['humidity']

# Применение функции к каждой строке основного DataFrame
main_df.apply(add_monthly_features, axis=1)

# Вывод основного DataFrame с добавленными фичами
#print(main_df.head())
```

```python
main_df.to_csv('data_with_climate.csv', index=False)
```
