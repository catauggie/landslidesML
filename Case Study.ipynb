```python
import osmnx as ox
import geopandas as gpd
import numpy as np
import pandas as pd
from shapely.geometry import box
import logging

# Define the geographic bounds
latitude_min = -18.3
latitude_max = -17.3
longitude_min = 177.2
longitude_max = 178.7


# Input number of sectors
x = 10#int(input("Enter the number of sectors along longitude: "))
y = 10#int(input("Enter the number of sectors along latitude: "))

logging.info("Fetching water body data from OpenStreetMap...")

logging.info("Fetching water body data from OpenStreetMap...")

# Fetch water body data from OSM using updated functions
try:
    place_name = 'Fiji'
    water_gdf = ox.features_from_place(place_name, tags={'natural': 'water'})
    water_gdf = gpd.GeoDataFrame(water_gdf, crs="EPSG:4326")  # Ensure CRS is WGS84
    logging.info("Successfully fetched water body data.")
except Exception as e:
    logging.error(f"Error fetching water body data: {e}")
    raise

logging.info("Generating grid sectors...")

# Generate grid coordinates
lat_lines = np.linspace(latitude_min, latitude_max, y + 1)
lon_lines = np.linspace(longitude_min, longitude_max, x + 1)

# Create a GeoDataFrame for grid sectors
grid_gdf = gpd.GeoDataFrame(geometry=[])

# Generate grid polygons
for lat in range(len(lat_lines) - 1):
    for lon in range(len(lon_lines) - 1):
        polygon = box(lon_lines[lon], lat_lines[lat], lon_lines[lon + 1], lat_lines[lat + 1])
        grid_gdf = grid_gdf.append({'geometry': polygon}, ignore_index=True)

# Ensure the CRS is consistent
grid_gdf.crs = water_gdf.crs

logging.info("Calculating water surface percentage in each grid sector...")

# Calculate water surface area percentage in each grid sector
results = []
for idx, sector in grid_gdf.iterrows():
    sector_geom = sector['geometry']
    
    # Intersection of the sector with the water surface areas
    try:
        intersected = water_gdf[water_gdf.intersects(sector_geom)]
        if intersected.empty:
            logging.warning(f"Sector {idx + 1} has no intersected water bodies.")
            water_area = 0
        else:
            water_area = intersected.geometry.unary_union.area
    except Exception as e:
        logging.error(f"Error processing sector {idx + 1}: {e}")
        water_area = 0
    
    sector_area = sector_geom.area
    
    # Calculate percentage of water surface
    water_percentage = (water_area / sector_area) * 100 if sector_area > 0 else 0
    
    # Calculate sector center coordinates
    sector_center = sector_geom.centroid
    center_lat = sector_center.y
    center_lon = sector_center.x
    
    # Append results
    results.append({
        'Sector_ID': idx + 1,
        'Center_Latitude': center_lat,
        'Center_Longitude': center_lon,
        'Water_Surface_Percentage': water_percentage
    })

    logging.info(f"Processed Sector {idx + 1}: Water Surface Percentage = {water_percentage:.2f}%")

# Convert results to DataFrame
results_df = pd.DataFrame(results)

logging.info("Saving results to Excel file...")

# Save results to Excel
try:
    results_df.to_excel('sector_water_surface.xlsx', index=False)
    logging.info("Results successfully saved to sector_water_surface.xlsx")
except Exception as e:
    logging.error(f"Error saving results to Excel file: {e}")
    raise
```

```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
from shapely.geometry import box, Point
import osmnx as ox

sector_df = pd.read_excel('sector_water_surface_20.xlsx')

# Ensure the columns are treated as strings before replacing commas
sector_df['Min_Latitude'] = sector_df['Min_Latitude'].astype(str).str.replace(',', '.').astype(float)
sector_df['Max_Latitude'] = sector_df['Max_Latitude'].astype(str).str.replace(',', '.').astype(float)
sector_df['Min_Longitude'] = sector_df['Min_Longitude'].astype(str).str.replace(',', '.').astype(float)
sector_df['Max_Longitude'] = sector_df['Max_Longitude'].astype(str).str.replace(',', '.').astype(float)
sector_df['Center_Latitude'] = sector_df['Center_Latitude'].astype(str).str.replace(',', '.').astype(float)
sector_df['Center_Longitude'] = sector_df['Center_Longitude'].astype(str).str.replace(',', '.').astype(float)


# Load point data (assuming fiji_df is provided)
fiji_df = pd.read_csv('fiji_coordinates.csv')  # Replace with actual data source


fiji_gdf = gpd.GeoDataFrame(
    fiji_df, 
    geometry=gpd.points_from_xy(fiji_df.longitude, fiji_df.latitude),
    crs="EPSG:4326"
)

# Create a GeoDataFrame for sectors
sectors = []
for idx, row in sector_df.iterrows():
    polygon = box(row['Min_Longitude'], row['Min_Latitude'], row['Max_Longitude'], row['Max_Latitude'])
    sectors.append({
        'Sector_ID': row['Sector_ID'],
        'geometry': polygon,
        'Water_Surface_Percentage': row['Water_Surface_Percentage'],
        'Center_Latitude': row['Center_Latitude'],
        'Center_Longitude': row['Center_Longitude']
    })

sectors_gdf = gpd.GeoDataFrame(sectors, crs="EPSG:4326")

# Initialize the map
place_name = "Fiji"
gdf = ox.geocode_to_gdf(place_name)
fig, ax = plt.subplots(figsize=(12, 12))

# Plot the OSM map
gdf.plot(ax=ax, color='white', edgecolor='black')

# Plot sectors with appropriate colors
for idx, sector in sectors_gdf.iterrows():
    color = 'green'
    if sector['Water_Surface_Percentage'] > 0:
        # Check if any points from fiji_df are within this sector
        if fiji_gdf.within(sector['geometry']).any():
            color = 'red'
    ax.add_patch(plt.Polygon(list(sector['geometry'].exterior.coords), closed=True, facecolor=color, edgecolor='black', alpha=0.3))

# Set axis limits and labels
ax.set_xlim(longitude_min, longitude_max)
ax.set_ylim(latitude_min, latitude_max)
ax.set_xlabel("Longitude")
ax.set_ylabel("Latitude")
ax.set_title("Sectors with Water Surface Percentage")

plt.show()
```
